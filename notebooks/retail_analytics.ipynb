## 6A - RFM Analysis

### Metodologi
- Recency: hari sejak transaksi terakhir (reference date: 2011-12-10)
- Frequency: jumlah invoice unik per customer
- Monetary: total revenue per customer
- Scoring: quintile 1–5 per dimensi (5 = terbaik)

### Segmen yang Dihasilkan
| Segmen | Jumlah | Avg Recency | Avg Frequency | Avg Monetary |
|---|---|---|---|---|
| Champions | 1.297 | 19.96 hari | 17.11x | £9.143 |
| Loyal Customers | 1.138 | 73.38 hari | 5.95x | £2.255 |
| Potential Loyalists | 358 | 79.72 hari | 2.70x | £591 |
| New Customers | 443 | 28.11 hari | 1.46x | £885 |
| Need Attention | 702 | 343.74 hari | 1.60x | £437 |
| Hibernating | 811 | - | - | - |
| At Risk | 616 | 359.51 hari | 5.62x | £4.244 |
| Lost | 513 | 568.23 hari | 1.00x | £344 |

### Key Insights
- **Champions (22.1%)** mendominasi revenue dengan Avg Monetary £9.143 -
  hampir 3x lipat segmen Loyal Customers
- **At Risk (10.5%)** adalah prioritas retensi utama - pernah bertransaksi
  besar (£4.244) tapi sudah 359 hari tidak aktif
- **Need Attention + Hibernating + Lost = 34.3%** dari total pelanggan -
  lebih dari sepertiga pelanggan sudah tidak aktif
- **New Customers (7.5%)** memiliki Avg Monetary £885 meski baru bergabung -
  potensi untuk di-nurture menjadi Loyal Customers

---

## 6B - K-Means Clustering

### Metodologi
- Fitur: Recency, Frequency, Monetary
- Preprocessing: Log Transformation (np.log1p) + StandardScaler
- Alasan Log Transformation: distribusi RFM sangat right-skewed,
  outlier wholesaler ekstrem dapat mendistorsi hasil clustering
- Jumlah cluster optimal: k=4 berdasarkan Elbow Method
- Variance explained oleh PCA: PC1 76.4% + PC2 18.7% = 95.1%

### Profil Cluster
| Cluster | Jumlah | Avg Recency | Avg Frequency | Avg Monetary |
|---|---|---|---|---|
| High Value | 1.196 | 27.71 hari | 19.28x | £10.731 |
| Promising | 1.250 | 28.30 hari | 3.05x | £857 |
| At Risk | 1.459 | 230.07 hari | 5.06x | £1.948 |
| Inactive | 1.973 | 394.95 hari | 1.38x | £317 |

### Key Insights
- **High Value & Promising memiliki Recency hampir sama** (27–28 hari) -
  yang membedakan adalah Frequency (19.3x vs 3.0x) dan
  Monetary (£10.731 vs £857), menunjukkan clustering berhasil
  memisahkan dua profil aktif yang berbeda
- **Inactive adalah cluster terbesar (1.973 customers)** - lebih dari
  33% pelanggan sudah tidak aktif lebih dari 1 tahun
- **At Risk masih memiliki nilai historis yang signifikan (£1.948)** -
  layak untuk diprioritaskan dalam win-back campaign
- **PCA menjelaskan 95.1% variance** - visualisasi 2D sudah sangat
  representatif untuk menggambarkan seluruh data

---

## 6C - Cohort Analysis

### Metodologi
- Pelanggan dikelompokkan berdasarkan bulan pertama bertransaksi
- Retention rate = % pelanggan yang kembali bertransaksi di bulan ke-N
- Total cohort: 25 bulan (Desember 2009 - Desember 2011)

### Key Insights
- **Retention bulan ke-1 berkisar 15–35%** - artinya 65–85% pelanggan
  tidak kembali di bulan berikutnya, umum untuk bisnis wholesale online
- **Cohort Desember 2009 paling loyal** - retention stabil di 25–45%
  hingga bulan ke-24, mengindikasikan pelanggan inti yang established
- **Peak season effect terlihat di diagonal heatmap** - nilai retention
  melonjak di bulan Oktober–November lintas cohort, membuktikan
  peak season mendorong returning customers bukan hanya new customers
- **Cohort Desember = one-time buyers** - retention bulan ke-1 hanya
  9.2% (2010-12), pelanggan holiday season cenderung tidak kembali
- **Mid-year cohort (Juli–Agustus) lebih loyal** dibanding cohort
  peak season - kemungkinan wholesaler yang mencari supplier rutin
- **Q3–Q4 2011 menunjukkan peningkatan retention** - kualitas akuisisi
  pelanggan baru yang semakin membaik menjelang akhir periode dataset

---

## Rekomendasi Bisnis

| Temuan | Segmen Target | Rekomendasi |
|---|---|---|
| At Risk punya nilai historis tinggi | At Risk (RFM & Cluster) | Win-back campaign dengan penawaran eksklusif |
| Champions = 22% customers, revenue terbesar | Champions | Loyalty program & early access produk baru |
| Desember = one-time buyers | New Customers (cohort Des) | Email onboarding series bulan Januari |
| Peak season mendorong returning customers | Semua segmen | Re-engagement campaign mulai September |
| Mid-year cohort paling loyal | - | Fokus akuisisi pelanggan di Q2–Q3 |
| Hibernating masih bisa di-recover | Hibernating | Reactivation campaign dengan diskon |

---

## Output Files
- `data/cleaned/rfm_segments.csv` ← RFM score, segmen & cluster label
- `data/cleaned/cluster_profile.csv` ← profil rata-rata per cluster
- `data/cleaned/cohort_retention.csv` ← retention matrix
- `reports/08_rfm_segments.png`
- `reports/09_elbow_method.png`
- `reports/10_cluster_pca.png`
- `reports/11_cluster_profile.png`
- `reports/12_cohort_heatmap.png`

````python
# Load hasil Phase 6
rfm = pd.read_csv('../data/cleaned/rfm_segments.csv')
cluster_profile = pd.read_csv('../data/cleaned/cluster_profile.csv')
cohort = pd.read_csv('../data/cleaned/cohort_retention.csv', index_col=0)

print("=== PHASE 6 SUMMARY ===")
print(f"\n📊 RFM Segmentation:")
for segment, count in rfm['Segment'].value_counts().items():
    pct = count / len(rfm) * 100
    print(f"  {segment:<25} : {count:>5} customers ({pct:.1f}%)")

print(f"\n🎯 K-Means Clustering:")
for _, row in cluster_profile.sort_values('Avg_Monetary', ascending=False).iterrows():
    print(f"  {row['Cluster_Label']:<15} : "
          f"R={row['Avg_Recency']:.0f}d | "
          f"F={row['Avg_Frequency']:.1f}x | "
          f"M=£{row['Avg_Monetary']:,.0f}")

print(f"\n📅 Cohort Analysis:")
print(f"  Total cohorts       : {len(cohort)} bulan")
print(f"  Avg retention M+1   : {cohort.iloc[:, 1].mean():.1f}%")
print(f"  Best cohort         : 2009-12 (retention stabil 25–45%)")
print(f"  Peak season effect  : Spike retention di Oct–Nov lintas cohort")

# Tampilkan key visualisasi
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, path, title in zip(axes,
    ['../reports/08_rfm_segments.png',
     '../reports/10_cluster_pca.png',
     '../reports/12_cohort_heatmap.png'],
    ['RFM Segmentation',
     'K-Means Clustering (PCA)',
     'Cohort Retention Heatmap']):
    img = mpimg.imread(path)
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(title, fontweight='bold')
plt.tight_layout()
plt.show()
````

---

## 7A - Revenue Analysis

### Key Insights
- **Total Revenue £17.374.804** dari 36.969 orders, 5.878 customers, dan
  4.631 produk dengan AOV £469.98
- **Pola seasonal sangat konsisten di kedua tahun** - revenue selalu

- **2010 Q4 (£2.769.995) vs 2011 Q4 (£2.709.056)** - meskipun terlihat

  Desember), artinya 2011 Q4 sebenarnya lebih besar jika data penuh
- **2011 Q1 (£1.608.268)** adalah satu-satunya kuartal yang turun YoY
  dibanding 2010 Q1 (£1.757.340)
- **AOV stabil di kisaran £400–£575** sepanjang periode - tidak ada
  perubahan signifikan dalam nilai per transaksi
- **Tren AOV H2 2011 meningkat** - nilai per transaksi semakin besar
  menjelang akhir periode, mengindikasikan pelanggan memesan dalam
  jumlah yang lebih besar

---

## 7B - Product Analysis

### Key Insights
- **REGENCY CAKESTAND 3 TIER (£277.656)** adalah produk revenue tertinggi
  dengan 3.317 orders - jauh di atas produk kedua WHITE HANGING HEART
  T-LIGHT HOLDER (£247.048)

  dari satu wholesaler, bukan indikator performa produk yang sebenarnya
- **Top 21.7% produk menghasilkan 80% revenue** - mendekati prinsip

- **~78% produk (4.100+ produk) hanya berkontribusi 20% revenue** -
  kandidat untuk evaluasi dan rasionalisasi portofolio produk
- **Bottom 10 produk memiliki revenue di bawah £25** meski sudah
  memiliki 10+ orders - harga per unit sangat rendah, perlu evaluasi

---

## 7C - Cancellation Analysis

### Key Insights
- **Cancellation rate 18.32% (orders)** - hampir 1 dari 5 orders
  dibatalkan, namun hanya **8.08% dari total revenue** - cancellation
  didominasi order bernilai kecil, bukan order besar
- **2.572 customers (43% dari total)** pernah melakukan cancellation
  setidaknya sekali
- **Tren cancellation rate menurun dari 2010 → 2011** - dari kisaran
  20–23% turun ke 14–17% di akhir 2011, mengindikasikan perbaikan
  operasional yang signifikan
- **Japan (41.1%)** memiliki cancellation rate tertinggi - lebih dari

- **Netherlands (8.4%)** adalah negara dengan cancellation rate terendah

- **11 dari 18 negara berada di bawah weighted average (18.3%)** -
  masalah cancellation terkonsentrasi di segelintir negara saja
- **REGENCY CAKESTAND 3 TIER (341 cancellations)** adalah produk
  paling sering dibatalkan sekaligus produk revenue tertinggi -
